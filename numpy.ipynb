```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

print(np.sum(a))          # suma total
print(np.mean(a))         # promedio
print(np.std(a))          # desviación estándar
print(np.var(a))          # varianza
print(np.min(a))          # mínimo
print(np.max(a))          # máximo
print(np.argmin(a))       # índice del valor mínimo
print(np.argmax(a))       # índice del valor máximo
```

```python
np.sum(a, axis=0)  # suma por columnas
np.sum(a, axis=1)  # suma por filas
```

```python
arr = np.array([[3, 1, 2],
                [6, 5, 4]])

print(np.sort(arr))           # ordena valores
print(np.argsort(arr))        # índices que ordenarían el array
print(np.argmax(arr, axis=1)) # índice del valor máximo por fila
print(np.argmin(arr, axis=0)) # índice del valor mínimo por columna
```
